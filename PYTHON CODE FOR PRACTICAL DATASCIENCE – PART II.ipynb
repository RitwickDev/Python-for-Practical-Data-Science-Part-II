```python
#Create new column 'final-grades' showing classes "Pass" or "Fail". If a student scores more than 50% in their final exam(G3),he/she is considered PASS,
#Otherwise they are considered FAIL.
def define_grade(studentmat):
    # Create a list to store the data
    final_grades = []

    # For each row in the column,
    for row in studentmat['G3']:
        # if more than a value,
        if row >= (0.5 * studentmat['G3'].max()):
                    final_grades.append('Pass')
        # else, if more than a value,
        elif row < (0.5 * studentmat['G3'].max()):
            # Append a letter grade
            final_grades.append('Fail')   
    # Create a column from the list
    studentmat['final_grades'] = final_grades
    return studentmat

final_data = define_grade(studentmat)
final_data.head(5)
```

```python
final_data.drop(["school","age"], axis=1, inplace=True)
```

```python
# Unfortunately, sklearnâ€™s machine learning library does not support handling categorical data.
# for yes / no values:
x1 = {'yes': 1, 'no': 0}
final_data['schoolsup'] = final_data['schoolsup'].map(x1)
final_data['famsup'] = final_data['famsup'].map(x1)
final_data['paid'] = final_data['paid'].map(x1)
final_data['activities'] = final_data['activities'].map(x1)
final_data['nursery'] = final_data['nursery'].map(x1)
final_data['higher'] = final_data['higher'].map(x1)
final_data['internet'] = final_data['internet'].map(x1)
final_data['romantic'] = final_data['romantic'].map(x1)
```

```python
# map the sex data
x2 = {'F': 1, 'M': 0}
final_data['sex'] = final_data['sex'].map(x2)

# map the address data
x3 = {'U': 1, 'R': 0}
final_data['address'] = final_data['address'].map(x3)

# map the familiy size data
x4 = {'LE3': 1, 'GT3': 0}
final_data['famsize'] =final_data['famsize'].map(x4)

# map the parent's cohabitation status
x5 = {'T': 1, 'A': 0}
final_data['Pstatus'] = final_data['Pstatus'].map(x5)

# map the parent's job
x6 = {'teacher': 0, 'health': 1, 'services': 2,'at_home': 3,'other': 4}
final_data['Mjob'] = final_data['Mjob'].map(x6)

#map the father's job 
x7 = {'teacher': 0, 'health': 1, 'services': 2,'at_home': 3,'other': 4}
final_data['Fjob'] = final_data['Fjob'].map(x7)

# map the reason data
x8 = {'home': 0, 'reputation': 1, 'course': 2,'other': 3}
final_data['reason'] = final_data['reason'].map(x8)

# map the guardian data
x9 = {'mother': 0, 'father': 1, 'other': 2}
final_data['guardian'] = final_data['guardian'].map(x9)

# map the grades data
x10 = {'Fail': 0, 'Pass': 1}
final_data['final_grades'] = final_data['final_grades'].map(x10)
```
